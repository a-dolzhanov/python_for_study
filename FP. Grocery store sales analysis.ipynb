```python
# вимкнення повідомлень про попередження
warnings.filterwarnings("ignore")
```

```python
#створення параметру повної дати та часу
df['Date_time'] = df['Date'].astype(str) + ' ' + df['Time'].astype(str)
df['Date_time'] = df['Date_time'].astype('datetime64[ns]')
```

```python
# створення окремого стовпчика з годинами транзакцій
df["Hour"] = df["Date_time"].dt.hour
```

```python
# умовне розподілення по підгрупам часовим
data = df["Date_time"].dt.hour
time=[]
for i in data:
    if i>6 and i <= 13 :
        time.append('Morning')
    elif i> 13 and i <= 18:
        time.append('Afternoon') 
    elif i>= 18 and i < 24:
        time.append('Evening') 
    else:
        time.append('Night')

df["Day_time"] = time
```

```python
# Чисті дані, з якими будемо працювати
# підготовка для очищення від транзакцій з неправильними Net Bill Amount
df_cleaning_temp = df[['Receipt Number', 'Amount', 'Net Bill Amount']]

# підрахунок коректних значень
pivot_amount_correct = df.pivot_table(index=['Receipt Number'], values=['Amount'], aggfunc='sum')

# переіменування колонок з однаковими назвами для подальшого об'єднання двох таблиць
pivot_amount_correct.columns = [c+'_correct' for c in pivot_amount_correct.columns]

# скидання індексу
pivot_amount_correct = pivot_amount_correct.reset_index()

# поєднання двох таблиць для подальшого видалення розбіжностей
df = pd.merge(left=df, right=pivot_amount_correct, how='left', left_on='Receipt Number', right_on='Receipt Number')

# було рядків з розбіжностями
len(df[df["Net Bill Amount"]!=df["Amount_correct"]])

# транзакцій до видалення рядків з розбіжностями
df["Receipt Number"].nunique()

# видалення рядків з розбіжностями
df = df.drop(df[df["Net Bill Amount"]!=df["Amount_correct"]].index)

# видалення непотрібного стовпчику
df = df.drop("Amount_correct", axis=1)

# транзакцій після видалення рядків з розбіжностями
df["Receipt Number"].nunique()
```

```python
# the data seasonality
# 1) Перевірка на сезонність
df['TrMonth'] = df['Date_time'].apply(lambda x: dt.datetime(x.year, x.month,1))
month_sales = df.groupby(['TrMonth'])['QTY','Price','Gross Bill Amount'].mean()
month_sales.plot(subplots=True, figsize=(15, 5))
plt.tight_layout()
plt.show()

print("\nЧи є сезонність у продажах?")
print("\nТак, з отриманих результатів - бачимо, що можемо спостерігати сезонність в предноворічний сезон - у грудні. Це може бути пов'язано зі святами, такими як Різдво.")
print("Кількість придбаних товарів та витачені суми покупців у цей період зростають.")
print("\nВарто проводити промо-акції у цей період, так як покупці охочіше купують товари.")
```

## Перевірка кореляції між показниками Month, Age, Gross Bill Amount, Payment Mode, Gender, Hour

```python
# 8) Перевірка кореляції між показниками Month, Age, Gross Bill Amount, Payment Mode, Gender, Hour
df_heatmap = df[['Month', 'Age', 'Gross Bill Amount', 'Payment Mode', 'Gender',  'Hour']]
df_heatmap['Payment Mode'] = df_heatmap['Payment Mode'].map({'Card': 1, 'Cash': 0, 'Wallet': 0.5})
df_heatmap['Gender'] = df_heatmap['Gender'].map({'Female': 1, 'Male': 0})
plt.figure(figsize=(10, 8))
matrix = np.triu(df_heatmap.corr())
sns.heatmap(df_heatmap.corr(), annot=True, linewidth=.7, mask=matrix, cmap="PuBuGn");

print('\nПеревірка кореляції між показниками Month, Age, Gross Bill Amount, Payment Mode, Gender, Hour')
print('\nОтримали наступний результат:')
print('1) Month корелює тільки з Gross Bill Amount.')
print('2) Age ні з чим не корелює.')
print('3) Gross Bill Amount корелює тільки з Month. Та має невелику кореляцію з Payment Mode.')
print('4) Payment Mode корелює тільки з Gross Bill Amount.')
print('3) Gender ні з чим не корелює.')
print('4) Hour ні з чим не корелює.')
print('\nЯкий можемо зробити висновок, та з чим далі працювати: 1) додатково дослідити сезонний переіод перед новим роком; 2) детальніше пропрацювати різні варіанти оплати.')
```

## Які продукти мають найбільші продажі та в яких країнах?

```python
# 9) Які продукти мають найбільші продажі та в яких країнах?
# Підрахунок доходу від товарів
df_top_products = df.groupby(['Category', 'Items', 'Country'])['Amount'].sum()
print(df_top_products.sort_values(ascending=False).head(10))
# візуальне зображення топ 10 товарів по країнам
top_products = df_top_products.sort_values(ascending=False).head(10).reset_index()
sns.barplot(x='Country', y='Amount', hue='Items', data=top_products)
plt.title("Top 10 product sales by country")
plt.xlabel('Country')
plt.ylabel('Total amount')
plt.show()

print('\nЯкі продукти мають найбільші продажі та в яких країнах?')
print('\nЗ отриманого результату бачимо, що топ товарів, що принесли більше за всіх доходу є у Singapore та Philippines.')
print('Також, 9 з 10 товарів належать до категорії "Beaty".')
print('\nТож, варто приділити увагу розвитку цієї категорії товарів, так як вона згідно з цих даних, є найприбутковішою.')
```

## Чи є різниця між методами оплати в розірізі суми доходу за товари на розподілі годин покупок?
